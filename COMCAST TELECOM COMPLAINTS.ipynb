```python
#Adding the date and time coloumn
datetimeIndex=pd.DatetimeIndex
df_telecom_complaints["Date_month_year"]=pd.DatetimeIndex(df_telecom_complaints["Date_month_year"])
df_telecom_complaints.head()
```

```python
df_telecom_complaints["Date_month_year"].value_counts
```

```python
#Ploting Daily Frequencies
df_telecom_complaints["Date_month_year"].value_counts().plot(xlabel='Complaint_Date',ylabel='no_of_complaints',title='No_of_Complains_perDay',color='r')
plt.show()
```

```python
#PLotting Monthly Frequencies 
df_telecom_complaints.sort_values(by=["Date_month_year"],inplace=True)
df_telecom_complaints["Date_month_year"].dt.month_name().value_counts(sort=False).plot(xlabel="Months_of_Complains",ylabel="No_of_Complains",title="No_of_Complaints_Per_Month",color='g')
plt.show()
```

```python
df_telecom_complaints["Custom _Status"]=["Open" if Status=="Pending" else "Closed" for Status in df_telecom_complaints ["Status"]]
```

```python
df_telecom_complaints
```

```python
State_group=df_telecom_complaints.groupby(["State","Status"]).size().unstack().fillna(0)  # filling Zeros if data is not there
print(State_group)
```

```python
#Statewise status of complaints
State_group.plot(kind='bar', stacked=True,figsize=(20,20),grid=True,xlabel="State",ylabel="Open&ClosedStatus",title="state_wise_status_of_complaints")
plt.show()
```

```python
#State with Maximum Complaints
State_group["Total"]=State_group["Closed"]+State_group["Open"]
print("State with Maximum Complaints is",State_group["Total"].index.max(),":",State_group["Total"].max())
```

```python
State_group
State_group["unresolved%"]=(State_group["Open"]/State_group["Total"])*100
```

```python
#State has the highest percentage of unresolved complaints
State_group
print("State with highest % of unresolved Compliants is\n",State_group["unresolved%"].index.max(),":",State_group["unresolved%"].max())
```

```python
#The percentage of complaints resolved till, which were received through the Internet and customer care calls
Int_cust_data=df_telecom_complaints.groupby(["Received Via","Status"]).size().unstack().fillna(0)
```

```python
print(Int_cust_data)
```

```python
Int_cust_data["Total"]=Int_cust_data["Closed"]+Int_cust_data["Open"]
```

```python
print("percentage of complaints resolved till date,which were received through the Internet and customer care calls are:")
print(((Int_cust_data["Closed"].sum())/(Int_cust_data["Total"].sum()))*100)
```
